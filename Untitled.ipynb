```python
def plot_hist(hist,item):
    if item=='top-5-acuuracy':
        iteml='top-5-accuracy'
    else:
        iteml=item
    plt.plot(hist.history[item],label=iteml)
    plt.plot(hist.history["val_"+item],label="val_"+iteml)
    plt.xlabel("Epochs")
    plt.ylabel(item)
    plt.title("Train and Validation {} Over Epochs".format(item), fontsize=14)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_hist(history_vgg,'loss')
plot_hist(history_vgg,'top-5-acuuracy')
plot_hist(history_vgg,'Accuracy')
plot_hist(history_vgg,'roc_auc')
```

```python
plot_hist(history_inc,'loss')
plot_hist(history_inc,'top-5-acuuracy')
plot_hist(history_inc,'Accuracy')
plot_hist(history_inc,'roc_auc')
```

```python
plot_hist(history_xcp,'loss')
plot_hist(history_xcp,'top-5-acuuracy')
plot_hist(history_xcp,'Accuracy')
plot_hist(history_xcp,'roc_auc')
```

```python
plot_hist(history_eff,'loss')
plot_hist(history_eff,'top-5-acuuracy')
plot_hist(history_eff,'Accuracy')
plot_hist(history_eff,'roc_auc')
```
